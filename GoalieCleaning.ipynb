```python
def goalie_cleaner(g_file):
    goalies= g_file.T
    goalies= goalies.set_index(0)
    goalies= goalies.T
    goalies= goalies.set_index('Player')
    goalies= goalies.drop(columns=['Rk','GSAA','GA%-','QS%'])
    goalies= goalies.rename(columns={'SO':'SUT'})
    return(goalies)

def g_SOT_cleaner (g_file2):
    g= g_file2.set_index('Player')
    g= g.rename(columns={'Att.':'SOT_Att','Made':'SOT_Scored','Miss':'SOT_Saved','Pct.':'SOT_Pct'})
    return (g)
    
    #then goalies_0809= pd.merge(g0809,g_shootout_0809,how='inner', on='Player')
    #then add the year g['season']=200x
    
def fix_tot(file):
    tot= file.loc[(slice(None),'TOT'),:]
```

#               ***GOALIE DATA FOR 2008-2009 SEASON***

```python
g0809=pd.DataFrame()
g0809= pd.read_excel('./Stats/2008-09/2008-09goalies.xlsx', engine= 'openpyxl')

gs_0809=pd.DataFrame()
gs_0809= pd.read_excel('./Stats/2008-09/2008-09goalies_shootout.xlsx', engine= 'openpyxl')

g0809=goalie_cleaner(g0809)
gs_0809= g_SOT_cleaner(gs_0809)

goalies0809= pd.merge(g0809,gs_0809,how='left', on='Player')

goalies0809['Season']=2008
goalies0809= goalies0809.reset_index()
goalies0809= goalies0809.set_index(['Player', 'Season'])
goalies0809= goalies0809.drop(columns='Team')

goalies0809= goalies0809[goalies0809.index.get_level_values('Player') != 'Player']

replace_nulls= goalies0809[['SOT_Att','SOT_Scored','SOT_Saved','SOT_Pct']].fillna (value=0)

goalies0809[['SOT_Att','SOT_Scored','SOT_Saved','SOT_Pct']]= replace_nulls

goalies0809= goalies0809.rename(columns={'T/O':'TplusOTL','SV%':'SV_Pct'})

goalies0809=goalies0809.reset_index()

### remove duplicates due to trades
new_list=[]

uniqueplayers=goalies0809.Player.unique()

for player in uniqueplayers:
    x= goalies0809[goalies0809.Player == player]
    uni_t= x.Tm.unique()
    if 'TOT' in uni_t:
        x= x[x.Tm == 'TOT']
    new_list.append(x)
    
new_df= pd.concat(new_list)
goalies0809= new_df.reset_index(drop=True)

goalies0809
```

```python
missing = pd.concat([goalies0809.isnull().sum(), 100 * goalies0809.isnull().mean()], axis=1)
missing.columns=['Count', '%']
missing.sort_values(by='Count', ascending=False)
```

```python
goalies0809.to_csv('./Clean/goalies2008.csv')
```

```python
missing = pd.concat([goalies_1819.isnull().sum(), 100 * goalies_1819.isnull().mean()], axis=1)
missing.columns=['Count', '%']
missing.sort_values(by='Count', ascending=False)
```

```python

goalies_1819.to_csv('./Clean/goalies2018.csv')
```
