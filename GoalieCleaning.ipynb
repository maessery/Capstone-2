```python
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
import os
import seaborn as sns
from datetime import datetime
import fnmatch
import sys
import glob
```

```python
xl_list = glob.glob('./Stats\**\*.xlsx', 
                   recursive = True)
for file in xl_list:
    return (file)
```

```python
files_xlsx= [f for f in xl_list if f[-4:] == "xlsx"]
files_xlsx
```

```python
def goalie_cleaner(g_file):
    goalies= g_file.T
    goalies= goalies.set_index(0)
    goalies= goalies.T
    goalies= goalies.set_index('Player')
    goalies= goalies.drop(columns='Rk')
    goalies= goalies.rename(columns={'SO':'SUT'})
    return(goalies)

def g_SOT_cleaner (g_file2):
    g= g_file2.set_index('Player')
    g= g.rename(columns={'Att.':'SOT_Att','Made':'SOT_Scored','Miss':'SOT_Saved','Pct.':'SOT_Pct'})
    return (g)
    
    #then goalies_0809= pd.merge(g0809,g_shootout_0809,how='inner', on='Player')
    #then add the year g['season']=200x
```

#               ***GOALIE DATA FOR 2008-2009 SEASON***

```python
g0809=pd.DataFrame()
g0809= pd.read_excel('./Stats/2008-09/2008-09goalies.xlsx', engine= 'openpyxl')

gs_0809=pd.DataFrame()
gs_0809= pd.read_excel('./Stats/2008-09/2008-09goalies_shootout.xlsx', engine= 'openpyxl')

g0809=goalie_cleaner(g0809)
gs_0809= g_SOT_cleaner(gs_0809)

goalies_0809= pd.merge(g0809,gs_0809,how='inner', on='Player')

goalies_0809['Season']=2008
goalies_0809= goalies_0809.reset_index()
goalies_0809= goalies_0809.set_index(['Season','Player'])

goalies_0809
```
