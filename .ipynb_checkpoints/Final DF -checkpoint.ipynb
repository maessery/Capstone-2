**ORGANIZE MULTIPLE POSITIONS**

```python
#### organize this position matrix. make sure to consoliate ld/rd, rd, ld all under 'D'
#### make sure only columns are c, rw, lw, d, g
```

```python
y= cf_contracts.POS.str.split(',').str.join('|').str.get_dummies()

cols= ['C1','D1','LW1','RD1','RW1',
        'C2', 'D2','G','LD','LD/RD',
        'LW2', 'RD2','RW2']
y.columns= cols

y['C1']= y['C1'] + y['C2']

y['D1']=y['D1']+ y['RD1']+ y['D2']+ y['LD'] + y['LD/RD'] + y['RD2']
y.loc[(y.D1 ==2), 'D1'] = 1

y['LW1'] = y['LW1'] + y['LW2']

y['RW1']= y['RW1'] + y['RW2'] 

y= y.drop(columns =['C2','D2','RD1','LD','LD/RD','LW2','RD2','RW2'])

y= y.rename(columns={'C1':'Pos_C', 'D1':'Pos_D', 'LW1': 'Pos_LW', 'RW1': 'Pos_RW', 'G': 'Pos_G'})


updated_contracts= pd.merge(left=cf_contracts, right= y, how= 'outer', left_on=cf_contracts.index, right_on=y.index)

contracts_and_positions= updated_contracts[['Player','AGE', 'Pos_C','Pos_LW','Pos_RW', 'Pos_D',
                                            'Pos_G', 'Season', 'TYPE', 'LENGTH', 'AAV']]
```
