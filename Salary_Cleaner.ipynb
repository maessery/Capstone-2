```python
def ht_conv(height):
    
    ht= height.split("\'")
    feet= float(ht[0])
    b= ht[1]
    b=b[:-1]
    inches= float(b)
    neat= 12* feet + inches
    return neat
```

```python
salary= pd.DataFrame()
salary=pd.read_excel(capdata[0], engine='openpyxl')
salary=salary.drop(columns=['TEAM','GP','P','G','A','P/GP','+/-','Sh','Sh%','TOI','CAP HIT','EXPIRY','CLAUSE','CAP HIT %','SALARY','BASE SALARY','MINORS','S.BONUS','P.BONUS'])
salary= salary[salary.TYPE != 'Entry-Level']
salary= salary[salary.AAV != 0]


salary=salary.reset_index()
salary=salary.drop(columns='index')

salary.PLAYER = salary.PLAYER.str.split('.').str[1]

salary= salary.set_index(['PLAYER','Season'])

salary.DRAFTED = salary.DRAFTED.str.split('-').str[0]

salary=salary.reset_index()

salary= salary.rename(columns={'PLAYER':'Player', 'SF%': 'SF_Pct', 'CF%': 'CF_Pct', 'FF%' : 'FF_Pct', 'xGF%': 'xGF_Pct', 'Sv%': 'Sv_Pct'})

heights= salary["HEIGHT"].tolist()
new_hts=[]
for height in heights:
    new_hts.append(ht_conv(height))
salary['HEIGHT']= new_hts

undrafted= salary['DRAFTED'].fillna (value=300)
salary['DRAFTED']=undrafted


salary
```
