```python
import pandas as pd
import os
from datetime import datetime
import fnmatch
import sys
import glob

from IPython.display import display

pd.set_option('display.max_columns', None)
```

```python
xl_list = glob.glob('./Stats\**\*.xlsx', recursive = True)

player_files =[]

for file in xl_list:
    player_files.append(file)

files_xlsx= [f for f in xl_list if f[-4:] == "xlsx"]
#files_xlsx
```

```python
player_files
```

```python
skaters=[]

for file in player_files:
    if file.endswith('skaters.xlsx') or file.endswith('misc.xlsx') or file.endswith('adv.xlsx') or file.endswith('skaters_shootout.xlsx'):
        skaters.append(file)
```

```python
skaters
```

```python
def preclean(file):
    file=file.drop(columns='Unnamed: 0')
    file=file.T
    file=file.set_index(0)
    file=file.T
    return (file)
    
def shootout_cleaner(file):
    file=preclean(file)
    file=file.drop(1, axis=0)
    file=file[['Player','Att.','Made','Miss','Pct.']]
    return (file)
    
    
```

#                  2008-2009 SEASON PLAYER STATS

```python
#load
s0809=pd.DataFrame()
s0809= pd.read_excel('./Stats/2008-09/2008-09skaters.xlsx', engine= 'openpyxl')

sa_0809=pd.DataFrame()
sa_0809=pd.read_excel('./Stats/2008-09/2008-09skaters_adv.xlsx', engine='openpyxl')

ss_0809=pd.DataFrame()
ss_0809=pd.read_excel('./Stats/2008-09/2008-09skaters_shootout.xlsx', engine='openpyxl')


#regular stats
s0809= preclean(s0809)
s0809['Season']=2008
s0809= s0809.set_index(['Player','Season'])



#advanced

sa_0809 = preclean(sa_0809)


sa_0809['Season']=2008
sa_0809= sa_0809.set_index(['Player','Season'])
sa_0809= sa_0809.drop(columns='Tm')



#shootout

ss_0809=ss_0809.reset_index()

ss_0809['Season']=2008

ss_0809=ss_0809.set_index(['Player','Season'])

ss_0809= ss_0809.drop(columns=['Tm','index'])

ss_0809= ss_0809.rename(columns={'Att.':'SOT_Att','Made':'SOT_Scored','Miss':'SOT_Saved','Pct.':'SOT_Pct'})


#merge and clean

skaters0809= s0809.merge(sa_0809, how='inner', on=['Player','Season']).merge(ss_0809, how='left', on=['Player','Season'])


skaters0809= skaters0809[skaters0809.index.get_level_values('Player') != 'Player']


skaters0809= skaters0809.fillna (value=0)



skaters0809= skaters0809.drop(columns=['Age_y','Pos_y','GP_y'])

skaters0809= skaters0809.rename(columns={'Age_x': 'Age', 'Pos_x': 'Pos', 'GP_x': 'GP',
                             skaters0809.columns[10]: 'EV_G', skaters0809.columns[7]: 'PlusMinus',
                             skaters0809.columns[11]: 'PP_G', skaters0809.columns[12]: 'SH_G', 
                             skaters0809.columns[14]: 'EV_A', skaters0809.columns[15]: 'PP_A', 
                             skaters0809.columns[16]: 'SH_A', skaters0809.columns[18]: 'S_Pct', 
                             'FO%': 'FO_Pct', 'CF%': 'CF_Pct', 'CF% rel': 'CF_Pct_rel', 
                             'FF%': 'FF_Pct', 'FF% rel': 'FF_Pct_rel', 'oiSH%': 'oiSH_Pct',
                             'oiSV%': 'oiSV_Pct', 'oZS%': 'oZS_Pct', 'dZS%': 'dZS_Pct',
                             'Thru%': 'Thru_Pct'})

skaters0809=skaters0809.dropna(subset=['PDO', 'oiSH_Pct', 'oiSV_Pct'])

skaters0809=skaters0809.reset_index()

### remove multiple rows for same player due to trades
new_list=[]

uniqueplayers=skaters0809.Player.unique()

for player in uniqueplayers:
    x= skaters0809[skaters0809.Player == player]
    uni_t= x.Tm.unique()
    if 'TOT' in uni_t:
        x= x[x.Tm == 'TOT']
    new_list.append(x)
    
new_df= pd.concat(new_list)
skaters0809 = new_df.reset_index(drop=True)

### remove duplicates
skaters0809= skaters0809.drop_duplicates(subset=['Player','Tm'], ignore_index=True)

skaters0809
```

```python
missing = pd.concat([skaters0809.isnull().sum(), 100 * skaters0809.isnull().mean()], axis=1)
missing.columns=['Count', '%']
missing.sort_values(by='Count', ascending=False)
```

```python
skaters0809.to_csv('./Clean/skaters2008.csv')
```

#                         2020-2021 PLAYER STATS

```python
#load
s2021=pd.DataFrame()
s2021= pd.read_excel('./Stats/2020-21/2020-21skaters.xlsx', engine= 'openpyxl')

sa_2021=pd.DataFrame()
sa_2021=pd.read_excel('./Stats/2020-21/2020-21skaters_adv.xlsx', engine='openpyxl')

ss_2021=pd.DataFrame()
ss_2021=pd.read_excel('./Stats/2020-21/2020-21skaters_shootout.xlsx', engine='openpyxl')


#regular stats
s2021= preclean(s2021)
s2021['Season']=2020
s2021= s2021.set_index(['Player','Season'])




#advanced

sa_2021 = preclean(sa_2021)


sa_2021['Season']=2020
sa_2021= sa_2021.set_index(['Player','Season'])
sa_2021= sa_2021.drop(columns='Tm')



#shootout

ss_2021=ss_2021.reset_index()

ss_2021['Season']=2020

ss_2021=ss_2021.set_index(['Player','Season'])

ss_2021= ss_2021.drop(columns=['Tm','index'])

ss_2021= ss_2021.rename(columns={'Att.':'SOT_Att','Made':'SOT_Scored','Miss':'SOT_Saved','Pct.':'SOT_Pct'})


# inner joined regular statistics with advance statistics because I only wanted to analyze players that are consistently on
# NHL rosters every night to get a better understanding of what a typcial player will make on his next salary (ie more than 10 
# games a season). Shootout stats were merged on with a left join because only a handful (relatively to the entire NHL
# population) of players participate in the shootouts; all null values were turned into 0's so that the machine learning
# model can take into a player's overall effectiveness in all aspects of the game, given how some players are acquired for
# sole purpose of being utilized in the shootout.

skaters2021= s2021.merge(sa_2021, how='inner', on=['Player','Season']).merge(ss_2021, how='left', on=['Player','Season'])


#clean some more

skaters2021= skaters2021[skaters2021.index.get_level_values('Player') != 'Player']


replace_nulls= skaters2021[['SOT_Att','SOT_Scored','SOT_Saved','SOT_Pct','FO%','S%','Thru%']].fillna (value=0)


skaters2021[['SOT_Att','SOT_Scored','SOT_Saved','SOT_Pct','FO%','S%','Thru%']]= replace_nulls


skaters2021= skaters2021.drop(columns=['Age_y','Pos_y','GP_y','E+/-'])

skaters2021= skaters2021.rename(columns={'Age_x': 'Age', 'Pos_x': 'Pos', 'GP_x': 'GP',
                             skaters2021.columns[10]: 'EV_G', skaters2021.columns[7]: 'PlusMinus',
                             skaters2021.columns[11]: 'PP_G', skaters2021.columns[12]: 'SH_G', 
                             skaters2021.columns[14]: 'EV_A', skaters2021.columns[15]: 'PP_A', 
                             skaters2021.columns[16]: 'SH_A', skaters2021.columns[18]: 'S_Pct', 
                             'FO%': 'FO_Pct', 'CF%': 'CF_Pct', 'CF% rel': 'CF_Pct_rel', 
                             'FF%': 'FF_Pct', 'FF% rel': 'FF_Pct_rel', 'oiSH%': 'oiSH_Pct',
                             'oiSV%': 'oiSV_Pct', 'oZS%': 'oZS_Pct', 'dZS%': 'dZS_Pct',
                             'Thru%': 'Thru_Pct', 'TOI/60': 'TOIper60', 'TOI(EV)': 'EV_TOI', 'SAtt.':'S_Att'})


### all players with null values were players with 1 GP, which means they did not play enough to accumulate significant
### stats in these columns. furthermore, this project is focused on more impactful players that are consistent NHLers

skaters2021=skaters2021.dropna(subset=['PDO', 'oiSH_Pct', 'oiSV_Pct'])

skaters2021=skaters2021.reset_index()

### remove multiple rows for same player due to trades
new_list=[]

uniqueplayers=skaters2021.Player.unique()

for player in uniqueplayers:
    x= skaters2021[skaters2021.Player == player]
    uni_t= x.Tm.unique()
    if 'TOT' in uni_t:
        x= x[x.Tm == 'TOT']
    new_list.append(x)
    
new_df= pd.concat(new_list)
skaters2021 = new_df.reset_index(drop=True)

### remove duplicates
skaters2021= skaters2021.drop_duplicates(subset=['Player','Tm'], ignore_index=True)

skaters2021
```
